```python
years = [1,2,3,4,3,2,3,]
salary = [12,12,14,18,13,13,11]

w = []

for i,j in zip(years,salary):
    w.append(j/i)

w = sum(w)/len(w)
w    
```

```python

for i,j in zip(years,salary):
    print(i,i*w)
```

```python
import matplotlib.pyplot as plt

years = [1,2,3,4,3,2,3,]
salary = [12,12,14,18,13,13,11]


plt.plot(years,salary)
plt.xlabel("years")
plt.ylabel("salary")
plt.title("salary vs years")
plt.show()
```
